### Lesson 3 : MAtplotlib
Numbers in a table are hard to understand at a glance. But a chart tells the story instantly. Matplotlib is Python's most popular visualization library — it turns your NumPy arrays and Pandas DataFrames into beautiful charts and graphs.

In ML, visualization is used to:

Explore data before training (spot patterns, outliers)
Evaluate models (plot accuracy, loss curves)
Present results (show predictions vs actual values)

import matplotlib.pyplot as plt → imports the plotting module, plt is the standard shortcut

plt.plot(x, y) → draws a line connecting x and y values

plt.title() → adds a title at the top

plt.xlabel() / plt.ylabel() → labels the axes

plt.show() → displays the plot — always the last line

***Styling***

color='red' → line color

linewidth=2 → thickness of line

linestyle='--' → dashed line ('-' = solid, ':' = dotted)

marker='o' → adds circle at each data point ('s' = square, '^' = triangle)

plt.grid(True) → adds background grid — easier to read values

Multiple plt.plot() calls add multiple lines to same chart

label= → name shown in legend

plt.legend() → displays the legend box showing which line is which

```python
from matplotlib.lines import lineStyles
import matplotlib.pyplot as plt
# Simple line plot
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 5]
# plt.plot(x, y)
# Styling the plot
plt.plot(x,y,color='red',linewidth=3,linestyle='--',marker='o')
plt.title('my first simple plot with styles ')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()

# multiple line plot
x = [1, 2, 3, 4, 5]
math    = [88, 72, 95, 60, 78]
science = [76, 85, 90, 55, 82]
plt.plot(x,math,color='purple',marker='s')
plt.plot(x,science,label='science',color='blue',marker='o')
plt.xlabel('Student')
plt.ylabel('Marks')
plt.title('Student Marks')
plt.legend()
plt.show()

```

### Concept 2: Bar Charts & Histograms

#### Short Theory — When to Use What?

Bar Chart → comparing categories (e.g. scores per student, sales per city)
plt.bar(x, y) → x = categories, y = values

Each category gets one bar

plt.barh() → horizontal bars — useful when category names are long

x and y labels swap compared to vertical bar chart

Histogram → showing distribution of values (e.g. how many students scored between 60-70?)

plt.hist(data, bins=5) → divides data into 5 equal ranges and counts values in each

bins → number of groups/buckets

edgecolor='black' → adds black border to each bar — easier to read


Both are used constantly in ML to understand your data before training.

### Concept 3: Scatter Plots & Subplots
Scatter Plot → shows relationship between two numerical variables. 

Subplots → multiple charts side by side in one figure — used to compare different visualizations at once

plt.scatter(x, y) → plots individual points, no connecting lines

s=100 → size of each point (bigger number = bigger dots)

marker='o' → shape of points

Pattern here → more hours studied = higher score = positive correlation

```python
import matplotlib.pyplot as plt
import numpy as np
#example 
hours_studied = [1, 2, 3, 4, 5, 6, 7, 8]
exam_scores   = [40, 50, 55, 65, 70, 80, 85, 95]
plt.scatter(hours_studied,exam_scores,color='red',marker='o',s=100)
plt.title('Hours vs Scores')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.grid(True)
plt.show()

#another example 
karachi_math = [88, 95, 78, 91]
karachi_sci  = [76, 90, 82, 88]
lahore_math  = [55, 42, 65, 70]
lahore_sci   = [85, 40, 70, 75]

plt.scatter(karachi_math, karachi_sci, color='blue',  label='Karachi', s=100)
plt.scatter(lahore_math,  lahore_sci,  color='red',   label='Lahore',  s=100)
plt.title('Math vs Science by City')
plt.xlabel('Math Score')
plt.ylabel('Science Score')
plt.legend()
plt.grid(True)
plt.show()
```

### using sub plots 
plt.subplots(1, 2) → 1 row, 2 columns = 2 charts side by side

figsize=(12, 4) → total figure width=12, height=4 inches

axes[0] → first chart, axes[1] → second chart

set_title(), set_xlabel() → same as plt.title() but for individual subplots

plt.tight_layout() → auto adjusts spacing so charts don't overlap

```python

import matplotlib.pyplot as plt
fig,axes=plt.subplots(1,2,figsize=(12,4))
#firtst chart
axes[0].plot(['Umar Khan.Ysfxi','ali','laiba'],[88,75,65],color='purple',marker='o',linewidth=2,linestyle='--')
axes[0].set_title('MAths score')
axes[0].set_xlabel('student')
axes[0].set_ylabel('student')
axes[0].grid(True)
# Second chart
axes[1].bar(['Umar', 'ali', 'laiba'], [76, 85, 90], color='green')
axes[1].set_title('Science Scores')
axes[1].set_xlabel('Student')
axes[1].set_ylabel('Score')
plt.tight_layout()
plt.show()
```

 ### 2x2 Subplots Grid means 4 charts on one page
 

```python
import matplotlib.pyplot as plt
fig,axes=plt.subplots(2,2,figsize=(12,8))
axes[0, 0].plot([1,2,3], [4,5,6],color='red',marker='o')
axes[0, 0].set_title('Top Left')
axes[0,0].grid(True)

axes[0, 1].bar(['A','B','C'], [3,7,5],color='purple')
axes[0, 1].set_title('Top Right')
axes[0,1].grid(True)

axes[1, 0].scatter([1,2,3], [3,1,4],color='brown',marker='^',s=100)
axes[1, 0].set_title('Bottom Left')
axes[1,0].grid(True)

axes[1, 1].hist([1,2,2,3,3,3,4])
axes[1, 1].set_title('Bottom Right',color='green')
axes[1,1].grid(True)

plt.tight_layout()
plt.show()
```
