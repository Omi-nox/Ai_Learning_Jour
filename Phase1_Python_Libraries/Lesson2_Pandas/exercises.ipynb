Same concept as NumPy boolean masking — just on a DataFrame now

# Adding, Updating & Deleting Columns
### CONCEPT 4

```python
import pandas as pd
data = {'Student': ['Umar', 'Ali', 'Sara', 'John'],
        'Math':    [88, 55, 95, 42],
        'Science': [76, 85, 90, 40]}
df=pd.DataFrame(data)
print(df)
#adding new col total
df['Total']=df['Math']+df['Science']  # ---->add values row by row (element-wise, just like NumPy)
print(df)
# new col added automatically to the right of data frame
# add passing col
df['Passing']=50
print(df)
#adding result col with conditions
df['Result']=df['Math']>=50 # it will give boolean
print(df)
#update an existing column by maping
df['Result']=df['Result'].map({True:'Pass',False:'Fail'})
print(df)
# updatin an column
df['Math']=df['Math']+5
print(df)
# changing student name john to jhon wick
df.loc[df['Student']=='John','Student']='John Wick'
print(f'method 1  : \n\n{df}')
df.loc[3,'Student']='John Wick'
print(f'method 2  : \n{df}\n')
print(df)
df.drop('Passing',axis=1,inplace=True) # inplace means do change and save in existing dataframe and do not create new one.
print(df)
# df = → must reassign! drop() doesn't modify original unless you add inplace=True
# RENAMING AN COL NAME
df.rename(columns={'Math':'Maths_SCORE'},inplace=True)
print(df)
```

### Detecting missing values
None → Python's way of representing missing data

Pandas converts it to NaN (Not a Number) internally

df.isnull() → returns boolean DataFrame — True = missing

df.isnull().sum() → counts missing values per column — most used

Row-wise check:

###Droping missing values
df.dropna() → removes every row that has at least one missing value

subset=['Math'] → only drops rows where Math is missing — other missing values kept-remove from specified col

Original df unchanged — must reassign

###Filling missing values
df.fillna(0) → replaces every NaN with 0

df['Math'].mean() → calculates mean ignoring NaN values automatically

Filling with mean is the most common strategy for numerical columns in ML

Filling with a label like 'Unknown' is common for text columns


```python
import pandas as pd
import numpy as np

data = {'Student': ['Umar', 'Ali', 'Sara', 'John', 'Ayesha'],
        'Math':    [88, None, 95, 42, 78],
        'Science': [76, 85, None, 40, 82],
        'Grade':   ['A', 'C', 'A', None, 'B']}

df = pd.DataFrame(data)
print(f'original data {df}')
print(df.isnull(),'\n')        # True where value is missing
print(df.isnull().sum(),'\n')  # count of missing values per column
dro=df.dropna() # drop every row where missing values are
print(dro,'\n')
nan_indexes = df[df.isnull().any(axis=1)] # make list of rows where nan exist and return their indexes
print(' rows with complete data having nan : ',nan_indexes,'\n')
nan_indexes = df[df.isnull().any(axis=1)].index.tolist() # make list of rows where nan exist and return their indexes
print('list of rows with index having nan : ',nan_indexes,'\n')
# gives names of specific col
cols_with_nan = df.columns[df.isnull().any()].tolist()
print('gives names of cols having nan',cols_with_nan,'\n')
#give specific cols list where its rows contain nan
nan_ids = df[df.isnull().any(axis=1)]['Math'].tolist()
print('give spcific col list having nan  :',nan_ids,'\n')
print(f'before dropping maths rows \n {df}  \n')
dro1=df.dropna(subset=['Math']) # drop row from maths col where missing values are
print(f'after dropping  maths rows\n {dro1} \n')

#FILLING MISSING VALUES
df1=df.fillna(0) # fill missing values with 0
print(f'filling with 0 \n{df1}\n')
df['Math']=df['Math'].fillna(df['Math'].mean()) # fill missing values with mean
print(f'filling with  mean maths col\n{df}\n')
df['Grade']=df['Grade'].fillna('Unknown') # fill missing values with Unknown
print(f'filling with Unknown Grade col \n{df}\n')





```

## Concept 6: GroupBy & Aggregation
In real ML projects you rarely look at individual rows. You ask questions like — "what's the average salary per department?", "which country scores the most goals?", "what's total sales per product category?" GroupBy lets you split data into groups and compute statistics per group — in one line.

df.groupby('Country') → splits DataFrame into groups based on unique Country values

['Goals'] → selects Goals column from each group

.mean() → computes average Goals for each group

Result is a Series with group names as index

.agg([...]) → apply multiple aggregation functions at once

Returns a DataFrame with one column per aggregation function

mean, sum, max, min are the most common ones

[['Goals', 'Assists']] → select multiple columns to aggregate
Returns a DataFrame with one row per group

```python
import pandas as pd
data = {'Player':  ['Ronaldo', 'Messi', 'Neymar', 'Mbappe', 'Salah', 'Benzema'],
        'Country': ['Europe', 'South America', 'South America', 'Europe', 'Africa', 'Europe'],
        'Goals':   [25, 30, 18, 22, 28, 20],
        'Assists': [10, 15, 12, 18, 9, 11]}

df = pd.DataFrame(data)

print(df,'\n')
df1=df.groupby('Country')['Goals'].mean()
df2=df.groupby('Country')[['Goals','Assists']].mean()
print('Basic Group by \n',df1,'\n\nGroupBy with Multiple Columns Aggregated\n',df2)
data= {'Department': ['IT', 'IT', 'HR', 'HR', 'IT'],
         'Gender':     ['Male', 'Female', 'Male', 'Female', 'Male'],
         'Salary':     [70000, 65000, 50000, 55000, 80000]}

df3 = pd.DataFrame(data)
print('\n',df3,'\n')
df4=df3.groupby(['Department','Gender'])['Salary'].agg(['mean','max','min','sum','count'])
print('\nmultiple aggregations\n',df4,'\n')
#
```

### CONCEPT 8 : MERGING AND JOINING
In real ML projects, data is rarely in one single table. You might have a students table with personal info and a separate scores table with exam results. To analyze them together you need to merge them. This is one of the most used operations in data engineering and ML preprocessing.

pd.merge(left, right, on='ID') → joins two DataFrames where ID matches

on='ID' → the common column to match rows on
 #### Types of join
*** Left join — keep all rows from LEFT table ***

left = pd.merge(students, scores, on='ID', how='left')

* Right join — keep all rows from RIGHT table **bold text**

right = pd.merge(students, scores, on='ID', how='right')

*** Outer join — keep ALL rows from BOTH tables***

outer = pd.merge(students, scores, on='ID', how='outer')

pd.concat([df1, df2]) → stacks df2 rows below df1

Simple rule: same columns, different rows → concat. Different columns, common key → merge.

ignore_index=True → resets index from 0 — without it index repeats 0,1,0,1

Used when you have same columns split across multiple files


```python
import pandas as pd
# Table 1 — Student info
students = pd.DataFrame({
    'ID':   [1, 2, 3, 4],
    'Name': ['Umar', 'Ali', 'Sara', 'John']})

# Table 2 — Student scores
scores = pd.DataFrame({
    'ID':    [1, 2, 3, 5],
    'Math':  [88, 55, 95, 70],
    'Science': [76, 85, 90, 60]})
mergerd=pd.merge(students,scores,on='ID') # the common thing will come to ones, as ID was common
print(mergerd)
# Default is inner join → only keeps rows where ID exists in both tables
# ID=4 (John) and ID=5 are dropped because they don't exist in both tables
```
