
### Lesson 2 Mini Project — Pandas Student Analysis System

This project uses every concept from Lesson 2 together. This is exactly the kind of data preprocessing work done before training any ML model.

Task 1 — Explore the Data:
Print shape, column names and missing value count per column.

Task 2 — Handle Missing Data:

Fill missing Math with mean of Math.

Fill missing Science with mean of Science. Fill missing English with mean of English.

Task 3 — Add New Columns:

Add Total = Math + Science + English

Add Average = Total / 3

Add Passed = True if Average >= 60, else False

Task 4 — Merge Tables:

Merge students with extra using left join on ID. How many rows have NaN in Grade column?

Task 5 — Filter Data:

From merged table, filter students who Passed AND paid their fee (Fee_Paid == True).

Task 6 — GroupBy:

Find average Math, Science and English scores per City.

Task 7 — Sort & Rank:
Sort students by Average score descending. Add a Rank column based on Average.

Task 8 — Final Summary:
Print the final cleaned and enriched DataFrame with all columns.

```python
import pandas as pd
import numpy as np


# Main student dataset
students = pd.DataFrame({
    'ID':      [1, 2, 3, 4, 5, 6, 7, 8],
    'Name':    ['Umar', 'Ali', 'Sara', 'John', 'Ayesha', 'Bilal', 'Zara', 'Omar'],
    'City':    ['Karachi', 'Lahore', 'Karachi', 'Lahore', 'Karachi', 'Lahore', 'Karachi', 'Lahore'],
    'Math':    [88, None, 95, 42, 78, 65, None, 70],
    'Science': [76, 85, None, 40, 82, 70, 88, 75],
    'English': [91, 68, 88, None, 79, 83, 85, 72]})

# Extra info table
extra = pd.DataFrame({
    'ID':       [1, 2, 3, 4, 5, 6, 9, 10],
    'Grade':    ['A', 'C', 'A', 'F', 'B', 'B', 'A', 'C'],
    'Fee_Paid': [True, False, True, False, True, True, False, True]})
df=pd.DataFrame(students)

# Task 1
print(f'Orgiginal data set 1 \n\n{students}\n and \nits extra info table \n\n{extra} ')
print(f'\nThe shape of the data : {students.shape}')
print(f'\nThe info of the whole data set  \n\n{students.info()}')
col_name=df.columns[df.any()].tolist()
print(f'The name of columns in the dataset \n {col_name}')
m3=df.isnull().sum()
print(f'the no. of missing values each columns {m3}')
mis_values=df[df.columns[df.isnull().any()]]
print('the cols that have nan \n\n',mis_values)
# Task 2
df['Math']=df['Math'].fillna(df['Math'].mean())
df['Science']=df['Science'].fillna(df['Science'].mean())
df['English']=df['English'].fillna(df['English'].mean())
print("columns fill with their mean \n\n",df)
# Task 3
df['Total']=df['Math']+df['Science']+df['English']
df['Average']=df['Total']/3
df['Passed']=df['Average']>=60
print(f'\nData set after updating \n\n{df}')
df['Passed']=df['Passed'].map({True:'Pass',False:'Fail'})
print(df)
#Task 4
df=pd.merge(students,extra,on='ID',how='left')
print(f'\nafter merging \n\n{df}')
gr_col=df[df.isnull().any(axis=1)]['Grade']
print('The grade col that have nan by rows are \n\n',gr_col)
print(f'the no of rows that have nan in grade col : {gr_col.isnull().sum()}')
#Task 5
unique_std=df[(df['Grade'].isin(['A','B','C']))&(df['Fee_Paid']==True)]
print(f'The students who are passed and paid their fee are \n\n{unique_std}')
grp_by=df.groupby('City')[['Math','Science','English']].mean()
print('avg of subjects for each city \n\n',grp_by)
#Task 6
df['Math']=df['Math'].fillna(df['Math'].mean())
df['Science']=df['Science'].fillna(df['Science'].mean())
df['English']=df['English'].fillna(df['English'].mean())
df['Total']=df['Math']+df['Science']+df['English']
df['Average']=df['Total']/3
df['Passed']=df['Average']>=60
df['Passed']=df['Passed'].map({True:'Pass',False:'Fail'})
#Task 7
print('\n\n after all updation the original data is \n\n',df)
rank=df['Average'].rank(ascending=False)
df['Rank']=rank
print(f'\n\nthe sorting of avg by descending{df.sort_values('Average',ascending=False).reset_index()}\n\n')
# print(f'adding ranking col\n\n',df)
#Task 8
print(f'final summary of data after all changes and updation \n\n{df}\n\nthe no. of cols is {df.shape[1]}')
```
